No Python, estruturas de controle como for, if e elif são fundamentais para controlar o fluxo de execução de um programa.

O for é usado para repetir um bloco de código para cada item de uma sequência (como uma lista, tupla, string ou range). Ele é ideal para iterações onde se sabe previamente o número de repetições.

A estrutura if permite a tomada de decisões. Ela executa um bloco de código somente se uma condição for verdadeira.

O elif (abreviação de "else if") é utilizado após um if e permite testar múltiplas condições. Se a condição do if for falsa, o programa verifica as condições dos blocos elif, em ordem, até encontrar uma que seja verdadeira.

Essas estruturas são essenciais para criar programas interativos e dinâmicos, permitindo desde simples verificações até laços de repetição complexos.

Apos decifrarmos o código temos que mostrar isso na tela ou imprimirmos como dizemos na linguagem do python, na **linha 20 a 23** vamos pedir ao python que imprima esses valores na tela primeiro vamos pedir que ele imprima a mensagem codificada **print("Mensagem codificada:", mensagem)** depois vamos pedir que ele imprima a mensagem decifrada **print("Mensagem decifrada :", mensagem_decifrada)** e para separar as mensagens vamos pedir que ele faça uma linha traçada entre as mensagens **print("-" * 40)**

```python
# Lista com mensagens codificadas
mensagens_codificadas = [
    "H3ll0! W0rLd#2025",
    "C0D1nG@Is#Fun!",
    "T3st@123WiTh#SyMb0ls"
]

# Iterar sobre cada mensagem na lista
for mensagem in mensagens_codificadas:
    mensagem_decifrada = ""  # Nova string para armazenar a mensagem decifrada

    # Loop aninhado para iterar sobre cada caractere da mensagem
    for caractere in mensagem:
        if caractere.islower():  # Se for letra minúscula
            mensagem_decifrada += caractere
        elif caractere.isupper():  # Se for letra maiúscula
            mensagem_decifrada += caractere.lower()
        # Números e símbolos são ignorados

    # Imprimir a mensagem original e a decifrada
    print("Mensagem codificada:", mensagem)
    print("Mensagem decifrada :", mensagem_decifrada)
    print("-" * 40)
```
